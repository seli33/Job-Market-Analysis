```python
df = pd.read_csv("data/raw/api_jobs_20260406.csv")
```

```python
df.isnull().sum()
```

```python
print("\n=== SAMPLE OF EMPTY ROWS ===")
print(df[df["salary_min"].isnull()][["title", "company", "salary_min"]].head(10))
```

```python
# Salary missing means not disclosed — don't drop, just label it
df["salary_min"] = df["salary_min"].fillna(0)
df["salary_max"] = df["salary_max"].fillna(0)

# Location missing — fill with Unknown
df["location"] = df["location"].fillna("Unknown")

# Remote missing — fill with False
df["is_remote"] = df["is_remote"].fillna(False)

# Description missing — fill with N/A
df["description"] = df["description"].fillna("N/A")
```

```python
df.isnull().sum()
```

```python
df["country"]=df["country"].fillna("unknown")
df["qualifications"] = df["qualifications"].fillna("N/A")
df["benefits"] = df["benefits"].fillna("Not Disclosed")
```

```python
df.isnull().sum()
```

```python
df.to_csv("data/processed/api_clean.csv",index=False)
print("Saved clean file")
```
